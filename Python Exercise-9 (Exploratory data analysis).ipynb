### Data Cleaning

```python
df.columns = df.columns.str.replace('"','')
```

```python
df.rename(columns=lambda x: x.strip(), inplace=True)
```

## 1. Which cities have the highest and lowest air quality?


```python
print('Cities have the highest air quality: ')
print('-'*120)
df[df['AirQuality'] == df['AirQuality'].max()]['City'].tolist()
```

```python
print('Cities have the lowest air quality: ')
print('-'*120)
df[df['AirQuality'] == df['AirQuality'].min()]['City'].tolist()
```

## 3. Are there any correlations between air quality and water pollution levels?


```python
print('Correlation between air quality and water pollution: ',df['AirQuality'].corr(df['WaterPollution']))
print('-'*120)
sns.heatmap(df[['AirQuality','WaterPollution']].corr(),annot=True)
plt.show()
```

## 4. Which cities have the highest and lowest water pollution levels?


```python
print('Cities have the highest WaterPollution: ')
print('-'*120)
df[df['AirQuality'] == df['WaterPollution'].max()]['City'].tolist()
```

```python
print('Cities have the lowest WaterPollution: ')
print('-'*120)
df[df['AirQuality'] == df['WaterPollution'].min()]['City'].tolist()
```
