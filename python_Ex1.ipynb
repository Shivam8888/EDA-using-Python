# If you want to see the Medal Trends Over the Years 
## Please refer below plot

```python
# Group the DataFrame by 'Year' and calculate the sum of medals for each year
df_grouped = df.groupby('Year')[['Gold Medals', 'Silver Medals', 'Bronze Medals']].sum().reset_index()

# Plot the trends
plt.plot(df_grouped['Year'], df_grouped['Gold Medals'], marker='o', label='Gold Medals')
plt.plot(df_grouped['Year'], df_grouped['Silver Medals'], marker='o', label='Silver Medals')
plt.plot(df_grouped['Year'], df_grouped['Bronze Medals'], marker='o', label='Bronze Medals')

# Set labels and title
plt.xlabel('Year')
plt.ylabel('Number of Medals')
plt.title('Medal Trends Over the Years')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
