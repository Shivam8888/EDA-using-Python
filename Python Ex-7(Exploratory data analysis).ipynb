```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import random
```

## 1. Plot a comparison plot of the following two expressions

### y = x^2 + 2x
### z = sin(y)

```python
# Define the range of x values
x = np.linspace(-4, 4, 80)

# Calculate y = x^2 + 2x
y = x**2 + 2*x

# Calculate z = sin(y)
z = np.sin(y)

# Create the comparison plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = x^2 + 2x')
plt.plot(x, z, label='z = sin(y)')
plt.xlabel('x')
plt.ylabel('y, z')
plt.title('Comparison Plot: y = x^2 + 2x vs. z = sin(y)')
plt.legend()
plt.grid(True)
plt.show()
```

## 3. Using subplot, demonstrate curve for sin(x), cos(x) and tan(x)

```python
fig, axes = plt.subplots(1, 3, figsize=(19, 6))
axes[0].plot(np.sin(x))
axes[1].plot(np.cos(x))
axes[2].plot(np.tan(x))
# Add a legend to each subplot
axes[0].legend()
axes[1].legend()

# Adjust spacing between subplots
plt.tight_layout()

# Display the subplots
plt.show()
```

## 4. Using the pie plot, demonstrate the contributions of runs stored by it's batsman in an innings(assume necessary data yourself)

```python
# Sample data for runs scored by each batsman
batsmen = ['Batsman A', 'Batsman B', 'Batsman C', 'Batsman D', 'Batsman E','Batsman F','Batsman G']
runs_scored = [78, 45, 32, 52, 63,20,80]

# Create the pie plot
plt.figure(figsize=(6, 6))
plt.pie(runs_scored, labels=batsmen, autopct='%1.1f%%', startangle=140)

# Set the aspect ratio to be equal to make the pie circular
plt.axis('equal')

# Add a title
plt.title('Contributions of Runs Scored by Batsmen in an Innings',pad=np.sort(runs_scored)[-1])

# Show the pie plot
plt.show()
```
