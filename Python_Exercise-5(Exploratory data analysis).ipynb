## I. What is the average life expectancy at birth for women of different countries?

```python
print('The average life expectancy at birth for women of different countries:')
print('-'*120)
for i in df['Country Name'].unique():
    data = df[df['Country Name'] == i]
    print(i,':' ,np.round(data['Health: Life expectancy at birth, female (years)'].mean(),2))
```

##  II. What is the correlation between the number of mobile phone subscribers and the number of internet users?

```python
df['Business: Mobile phone subscribers'] = df['Business: Mobile phone subscribers'].str.replace(',','')
df['Business: Mobile phone subscribers'] = df['Business: Mobile phone subscribers'].astype(int)
```

```python
print('Correlation between the number of mobile phone subscribers and the number of internet users: ',
      np.round(df['Business: Mobile phone subscribers'].corr(df['Business: Internet users (per 100 people)']),2))
print('-'*120)
sns.heatmap(df[['Business: Mobile phone subscribers','Business: Internet users (per 100 people)']].corr(),annot=True)
plt.show()
```

## III. What is the difference in the average GDP per capita between urban and rural areas?

```python
df['Finance: GDP per capita (current US$)'] = df['Finance: GDP per capita (current US$)'].str.replace(',','').astype(int)
df['Population: Total (count)'] = df['Population: Total (count)'].str.replace(',','').astype(int)
df['Population: Urban (count)'] = df['Population: Urban (count)'].str.replace(',','').astype(int)
df['Population: Rural (count)'] = df['Population: Total (count)'] - df['Population: Urban (count)']
```

```python
# Step 1: Calculate the average GDP per capita for urban and rural areas
average_gdp_urban = df.loc[df['Population: Urban (count)'] > 0, 'Finance: GDP per capita (current US$)'].mean()
average_gdp_rural = df.loc[df['Population: Rural (count)'] > 0, 'Finance: GDP per capita (current US$)'].mean()

# Step 2: Calculate the difference in the average GDP per capita between urban and rural areas
difference_gdp_per_capita = average_gdp_urban - average_gdp_rural

print("Difference in average GDP per capita between urban and rural areas:", difference_gdp_per_capita)
```

##  IV. Findtop five countries with the highest number of passenger cars per 1,000 people?

```python
data = df.groupby('Country Name').agg({'Transit: Passenger cars (per 1,000 people)':'sum'}).reset_index().sort_values('Transit: Passenger cars (per 1,000 people)',
                                                                                                               ascending=False).reset_index(drop=True)
```

## V. What is the relationship between the population's age distribution and the country's GDP per capita?

```python
# Calculate the average GDP per capita for the entire dataset
average_gdp_per_capita = df['Finance: GDP per capita (current US$)'].mean()

# Calculate the average GDP per capita for each age group
average_gdp_0_14 = df.loc[:, 'Finance: GDP per capita (current US$)'].mean() * df['Population: Ages 0-14 (% of total)'].mean() / 100
average_gdp_15_64 = df.loc[:, 'Finance: GDP per capita (current US$)'].mean() * df['Population: Ages 15-64 (% of total)'].mean() / 100
average_gdp_65_plus = df.loc[:, 'Finance: GDP per capita (current US$)'].mean() * df['Population: Ages 65+ (% of total)'].mean() / 100

# Print the results
print("Average GDP per capita for the entire dataset:", average_gdp_per_capita)
print("Average GDP per capita for ages 0-14:", average_gdp_0_14)
print("Average GDP per capita for ages 15-64:", average_gdp_15_64)
print("Average GDP per capita for ages 65+:", average_gdp_65_plus)
```

## VII. Name the countries with the highest and lowest GDP per capita?

```python
#Country with highest GDP per capita
df[df['Finance: GDP per capita (current US$)'] == df['Finance: GDP per capita (current US$)'].max()]['Country Name'].tolist()
```

```python
#Country with lowest GDP per capita
df[df['Finance: GDP per capita (current US$)'] == df['Finance: GDP per capita (current US$)'].min()]['Country Name'].tolist()
```

## VIII. How does the GDP vary among different countries?

```python
# Group the DataFrame by 'Country Name'
grouped_df = df.groupby('Country Name')

# Calculate the mean GDP per capita for each country
mean_gdp_per_capita_by_country = grouped_df['Finance: GDP per capita (current US$)'].mean()

# Display the mean GDP per capita for each country
print(mean_gdp_per_capita_by_country)
```

```python
plt.figure(figsize=(10, 37))
mean_gdp_per_capita_by_country.plot(kind='barh')
plt.xlabel('Mean GDP per capita')
plt.ylabel('Country Name')
plt.title('Mean GDP per capita by Country')
plt.xticks(rotation=90)
plt.show()
```
